```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Helvetica"
```

```python
plt.figure(figsize=(3,5),dpi=500)
# import data, raw numbers from histograms created with clusterfilterproc
# divide each by 3 because the sample had 3 events
nofilter=np.array([219469,186989,52031,45531,36392,31945,30188,28300])/3
loosef=np.array([120335,99176,24342,21133,21414,19019,21314,20055])/3
tightf=np.array([90587,73262,14176,11978,11615,10238,12032,11435])/3
DL = np.array([19427,19010,684,683,291,292,189,189])
# areas from old geo
areas=np.array([270.40,270.40,448.50,448.50,655.20,655.20,904.80,904.80])

# normalize to areas for hit density
nof = nofilter/areas
lf = loosef/areas
tf = tightf/areas
dlf = DL/areas

#plot
bins=np.arange(0,9) 
plt.hist(bins[:-1], bins, weights=nof,linewidth=1,color='#2799ff',edgecolor='black',histtype='stepfilled',label="90ps Timing Cut")
plt.hist(bins[:-1], bins, weights=lf,linewidth=1,color='#db8673',edgecolor='black',histtype='stepfilled',label="+ Loose Cluster Filter")
plt.hist(bins[:-1], bins, weights=tf,linewidth=1,color='#33cc99',edgecolor='black',histtype='stepfilled',label="+ Tight Cluster Filter")
plt.hist(bins[:-1], bins, weights=dlf,linewidth=1,color='#ff99cc',edgecolor='black',histtype='stepfilled',label="+ DL Filter")

# plot aesthetics
plt.yscale('log')
ticks = [1e-1,1,10,1e2,1e3,1e4]
xticks = np.arange(0,9)
plt.yticks(ticks)
#plt.xticks(xticks)
plt.legend()
plt.xlabel("Layer ID")
plt.xlim(0,8)
plt.ylabel(r"Hit Density (cm$^{-2}$)")
plt.title("Hit Density by Layer in Vertex Barrel Detector",y=1.02)
plt.grid(linestyle='dotted',alpha=1)
plt.tick_params('both',which='both',direction='in')

plt.savefig("/Users/chris/Desktop/hitdensity_mine.jpeg")
```
