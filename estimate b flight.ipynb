# quick estimates of distance travelled by a B hadron

```python
import numpy as np
import matplotlib.pyplot as plt
# from higgs decay, p = 60 GeV
bg_60= 12
# if p = 200 gev
bg_200 = 200/5
bg_500 = 500/5
c = 3e10 # cm/s
x = np.linspace(0,10,1000) #cm 
tau = 1.64e-12 #s
p_x_60 = np.exp(-x/(c*bg_60*tau))
p_x_200 = np.exp(-x/(c*bg_200*tau))
p_x_500 = np.exp(-x/(c*bg_500*tau))

# find some points of typical decays
h = [p_x_60,p_x_200,p_x_500]
pts = np.zeros(3)
thresh = 1/np.exp(1)
for i, f in enumerate(h):
    for ix in range(len(x)):
        if f[ix] < thresh: # if less than e^(-1)
            pts[i] = x[ix] 
            break
            
#plot
plt.figure(figsize=(6,4),dpi=500)
plt.plot(x,p_x_60,label=r"$p = 60$ GeV (ex: Higgs decay)")
plt.plot(x,p_x_200,label=r"$p = 200$ GeV")
plt.plot(x,p_x_500,label=r"$p = 500$ GeV")
plt.plot(pts,[thresh]*3,'o',label=r"$\Delta x$ for typical decay")
plt.xlabel(r"Distance travelled by B hadron (cm)")
plt.ylabel("Prob of travelling distance before decay")
plt.title("Possible flight distance of B hadrons before decay")
plt.legend()
plt.savefig("/Users/chris/Desktop/MC/plots/bflight.png")
```
